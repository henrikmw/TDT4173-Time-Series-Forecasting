```python
import pandas as pd
```

```python
gradient_boosted_result = pd.read_csv('gradient_boosted_submissions/gradient_boosted_1.csv')
random_forest_result = pd.read_csv('random_forest_submissions/random_forest_3.csv')
```

```python
# Calculate the average of the specified column from both DataFrames
average_column_df1 = gradient_boosted_result['prediction']
average_column_df2 = random_forest_result['prediction']

# Take the average of the two columns
average_column = (average_column_df1 + average_column_df2) / 2

print(average_column)

test = pd.read_csv('test.csv')
sample_submission = pd.read_csv('sample_submission.csv')

test['prediction'] = average_column
sample_submission = sample_submission[['id']].merge(test[['id', 'prediction']], on='id', how='left')
sample_submission.to_csv("ensemble_1.csv", index=False)
```
