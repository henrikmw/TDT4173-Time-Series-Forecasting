```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
mats = pd.read_csv('pred_59.csv')
catboost = pd.read_csv('catboost/catboost_1.csv')
```

```python
mats
```

```python
catboost
```

```python
# Calculate the average of the specified column from both DataFrames
average_column_df1 = mats['prediction']
average_column_df2 = catboost['prediction']

# Take the average of the two columns
average_column = (average_column_df1 + average_column_df2) / 2

print(average_column)

test = pd.read_csv('test.csv')
sample_submission = pd.read_csv('sample_submission.csv')

test['prediction'] = average_column
sample_submission = sample_submission[['id']].merge(test[['id', 'prediction']], on='id', how='left')
sample_submission.to_csv("ensemble_3.csv", index=False)
```
